Implement Conway’s Game of Life and visualize it with matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
# Make plots show up inline
#%matplotlib inline
# Make plots pop up
%matplotlib qt
```

```python
# Grid settings
rows, cols = 5, 5
# This is an object that can be used to generate random numbers. We will use it to create the initial grid.
rng = np.random.default_rng(0)
#Population Statistics (optional)
#grid = (rng.random((rows, cols)) < 0.35).astype(int)  # random 0/1 with ~35% alive
grid = rng.integers(0, 2, size=(rows, cols))  # random 0/1 with 50% chance each
grid
```

```python
has_one=np.where(grid==1)
coords = np.argwhere(grid == 1)
coords
```

```python
m_grid=np.zeros((rows,cols),dtype=int)
m_grid[tuple(coords.T)] = 7
m_grid
```

```python
def count_neighbors(p_grid):
    # Sum of 8 surrounding positions
    return (
        np.roll(p_grid,  1, 0) + np.roll(p_grid, -1, 0) +
        np.roll(p_grid,  1, 1) + np.roll(p_grid, -1, 1) +
        np.roll(np.roll(p_grid, 1, 0),  1, 1) +
        np.roll(np.roll(p_grid, 1, 0), -1, 1) +
        np.roll(np.roll(p_grid,-1, 0),  1, 1) +
        np.roll(np.roll(p_grid,-1, 0), -1, 1)
    )
```

```python
def step(p_grid):
    neighbors = count_neighbors(p_grid)
    return (
        ((p_grid == 1) & ((neighbors == 2) | (neighbors == 3))) |
        ((p_grid == 0) & (neighbors == 3))
    ).astype(int)
```

```python
import matplotlib.pyplot as plt

running = True

def on_key(event):
    global running
    running = False

plt.ion()
fig, ax = plt.subplots()
fig.canvas.mpl_connect("key_press_event", on_key)

while running:
    ax.clear()
    ax.imshow(grid, cmap="binary", interpolation="nearest")
    ax.axis("off")
    fig.canvas.draw()
    fig.canvas.flush_events()
    plt.pause(0.05)
    grid = step(grid)

plt.close(fig)
```
